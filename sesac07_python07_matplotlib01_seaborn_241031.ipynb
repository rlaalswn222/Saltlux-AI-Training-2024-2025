```python
import numpy as np
import pandas as pd

x= np.arange(0, 5, 0.1)
y1 = np.cos(x)
y2 = np.exp(x)
y3 = y1 * y2

plt.plot(x, y1)
plt.show()

plt.plot(x, y2)
plt.show()

plt.plot(x, y3)
plt.show()
```

```python
#subplot
#아예 고정 안함
x = np.arange(0, 5, 0.1)
y1 = np.cos(x)
y2 = np.exp(x)
y3 = y1 * y2

plt.subplot(131) #3행 1열 첫번째
plt.plot(x, y1)
plt.title('y = cos(x)')

plt.subplot(132) #3행 1열 두번째
plt.plot(x, y2)
plt.title('y = exp(x)')

plt.subplot(133) #3행 1열 세번째
plt.plot(x, y3)
plt.title('y = cos(x) x exp(x)')
```

```python
#세개의 그래프를 다같이 씀


x = np.arange(0, 5, 0.1)
y1 = np.cos(x)
y2 = np.exp(x)
y3 = y1 * y2

#plt.plot(x, y1, x, y2, x, y3)
plt.plot(x, y1, color='blue', label='y = cos(x)')
plt.plot(x, y2, color='green', label='y = exp(x)')
plt.plot(x, y3, color='red', label='y = cos(x) * exp(x)')
plt.legend(['y = cos(x)', 'y = exp(x)', 'y=cos(x)*exp(x)'])
```

```python
#fig.add_subplot()
#Figure에 직접 subplot만들어줌
#잘 안씀 나도 처음봄
#고정을 안시키고 나중에 내가 따로 고정해줌

x = np.arange(0, 5, 0.1)
y1 = np.cos(x)
y2 = np.exp(x)
y3 = y1 * y2

fig = plt.figure()

ax1 = fig.add_subplot(311)
ax1.plot(x, y1)
ax1.set_title('y = cos(x)')

ax2 = fig.add_subplot(312)
ax2.plot(x, y2)
ax2.set_title('y = exp(x)')

ax3 = fig.add_subplot(313)
ax3.plot(x, y3)
ax3.set_title('y = cos(x) * exp(x)')
```

```python
fig, axs = plt.subplots(3, 1) #행과 열을 먼저 고정 시키고 그림, 여기는 subplots에 s 붙음
#fig와 axs를 두개 받음
#fig.subplots_adjust(hspace=0.5) #그래프간 간격
fig.tight_layout() #더 자주 쓰이는 그래프 간 간격

axs[0].plot(x, y1)
axs[0].set_title('y = cos(x)')

axs[1].plot(x, y2)
axs[1].set_title('y = axp(x)')

axs[2].plot(x, y3)
axs[2].set_title('y = cos(x) * exp(x)')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 201)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x, y1, label = 'sin')
ax.plot(x, y2, label = 'cos')

ax.legend(loc=1)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.5)
y1 = x
y2 = x**2
y3 = np.sin(x)
y4 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2, marker = 'D')
plt.plot(x, y3, color='r')
plt.plot(x, y4, linestyle = 'dashed')
plt.show()
```

```python
plt.plot(x, y1)

plt.figure()
plt.plot(x, y2, marker = 'D')

plt.figure()
plt.plot(x, y3, color = 'r')

plt.figure()
plt.plot(x, y4, linestyle = 'dashed')
plt.show()
```

```python
import seaborn as sns

print("Seaborn version :", sns.__version__)
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x='total_bill', y='tip', hue='smoker', style='smoker', data = tips)
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value=np.random.randn(500).cumsum()))
g = sns.relplot(x='time', y='value', kind = 'line', data = df)
g.fig.autofmt_xdate()
```

```python
sns.catplot(x='day', y='total_bill', hue='smoker', col ='time', aspect=0.6, kind = "swarm", data = tips)
```

```python
titanic = sns.load_dataset('titanic')
g = sns.catplot(x='fare', y='survived', row = 'class',
                kind = 'box', orient ='h', height =1.5, aspect =4,
                data = titanic.query('fare>0'))

g.set(xscale='log')
```
