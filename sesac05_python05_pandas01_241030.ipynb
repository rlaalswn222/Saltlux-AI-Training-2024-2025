#apply

어떤 함수를 series의 각각의 element에 적용시켜주는 함수

map함수 보다 적용할 수 있는 함수의 범위가 넓음



```
seires.apply(func, args, **kwds)
```

func: series에 적용될 함수

args: series를 제외한 함수에 들어갈 다른 매개변수

**kwds: 함수에 넘겨줄 키워드 인자들


````python
s = pd.Series([20, 21, 12], index=['London', 'New York', 'Heisinki'])
s
````

````python
def sub_custom_value(x, val):
  return x - val

s.apply(sub_custom_value, args=(10,))
#val이라는 인자에 10을 전달
````

````python
#s.apply(sub_custom_value, args=10)
#TypeError: Value after * must be an iterable, not int
#apply메서드는 iterable 순환하게 생성되는데 변수는 iterable하지 않아서 오류가 나는거
````

````python
def add_custom_values(x, **kwargs): #kwargs는 딕셔너리 형식으로 키랑 value같이 쌍으로 들어감
  for month in kwargs:
    x += kwargs[month] #(kwargs[june]) = 30 + (kwargs[july] = 20) + (kwargs[august] =25) key값을 갖고 옴
  return x
````

````python
s.apply(add_custom_values, june=30, july=20, august=25)
````
