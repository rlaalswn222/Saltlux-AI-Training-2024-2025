````python
ages = [20, 25, 40 , 23, 14, 2, 4, 53, 12, 56, 24, 78, 45, 23, 13, 64, 42, 5,24,12,6, 68, 23]
bins = [0, 20, 40, 60, 80]
cuts = pd.cut(ages, bins)
cuts
````

````python
cuts.codes
````

````python
group_names = ['10대 이하', '20~30대', '40~50대', '60대 이상']
````

# get_dummies

명목형 변수를 one-hot-encoding

danse matrix: 속도는 빠르나, 관계나 순서 파악이 어려움

spase matrix: 매모리 많이 잡아 먹으나 관계나 순서 파악에 용이함



````python
df = pd.DataFrame({'col1': [10, 20, 30],
                   'col2': ['a', 'b', 'a']})
````

````python
df
````

````python
pd.get_dummies(df) #명목형 변수를 one-hot-encoding해줌 True=1, False=0
````

````python
pd.get_dummies(df, dtype = int)
````

#GroupBy
그룹 연산은 분리 - 적용 - 결합 과정



````python
kbo = pd.read_csv('/content/drive/MyDrive/kbo.csv')
kbo.head()
````

````python
kbo.shape
````

````python
kbo['팀'].unique()
````

````python
kbo.describe()
````

````python
kbo.info()
````

````python
kbo.groupby('팀').count() #넥센이 키움으로 바뀐걸 몰라서 이상함
````

````python
kbo.groupby(['연도','팀']).sum()
````

````python
kbo.groupby(['연도', '팀'])[['승률', '순위']].max()
````

````python
grouped = kbo.groupby('팀')
for name, group in grouped:
  print(name)
  print(group)

  print('-'*50)
````
